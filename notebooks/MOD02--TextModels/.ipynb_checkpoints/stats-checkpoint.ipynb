# Configure

```python
text_file = 'austen-persuasion.csv'
OHCO = 'chap_num para_num sent_num token_num'.split() # THESE MATCH HEADER IN CSV
```

# Set Up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

# Import Text

```python
text = pd.read_csv(text_file, index_col=OHCO)
```

```python
text = text[~text.token_str.isna()] # Remove NULLs
```

## Define general function to recreate OHCO chunks

```python
def gather_chunks(df, ohco, doc_str = 'token_str', sep=' '):
    chunks = df.groupby(ohco)[doc_str].apply(lambda x: x.str.cat(sep=sep)).to_frame().rename(columns={doc_str:'doc_content'})
    return chunks
```

# Create Vocabulary table

```python
text['term_str'] = text.token_str.str.lower().str.replace(r'[\W_]', '')
```

```python
vocab = text.term_str.value_counts().to_frame()\
    .reset_index()\
    .rename(columns={'index':'term_str', 'term_str':'n'})
vocab = vocab.sort_values('term_str').reset_index()
vocab.index.name = 'term_id'
vocab = vocab.drop('index', 1)
```

```python
vocab
```

## Show top 25 terms

```python
vocab.sort_values('n', ascending=False).head(25)
```

```python
top_terms = vocab.sort_values('n', ascending=False).head(25)
plt.figure(figsize=(5,10))
sns.barplot(data=top_terms, y='term_str', x='n', orient='h')
plt.show()
```

## Add term ID to tokens

```python
text['term_id'] = text['term_str'].map(vocab.reset_index().set_index('term_str').term_id).fillna(-1).astype('int')
```

```python
text.sample(10)
```

## Define convenience functions for vocabulary

```python
def get_term_id(term_str):
    try:
        return vocab[vocab.term_str == term_str].index[0]
    except IndexError:
        return None

def get_term_str(term_id):
    try:
        return vocab.loc[term_id].term_str
    except KeyError:
        return None
```

# Experiment -- Can We Detect Stopwords by Entropy?

```python
TDM = text[text.term_id >= 0].groupby(['chap_num','term_str']).term_str.count().to_frame().unstack().fillna(0.00001).T
```

```python
TDM = TDM.apply(lambda x: x / x.sum(), 1)
```

```python
TDH = TDM.apply(lambda x: -(x * np.log2(x))).sum(1).to_frame().rename(columns={0:'H'})
```

```python
TDH = TDH.reset_index().drop('level_0', 1)
```

```python
TDH.H.idxmax()
```

```python
TDH.loc[2540]
```

```python
TDH.sort_values('H', ascending=False)\
    .reset_index().set_index('term_str').H\
    .head(50).sort_values().plot(kind='barh', figsize=(10,20))
```

```python
## Add entropy to Vocabulary
```

```python
vocab = vocab.merge(TDH, on='term_str', copy=False)
```
